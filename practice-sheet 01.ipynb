Q5: A football is kicked 1 meter from the ground at an angle of 30 degree with a velocity of 15 m/sec. 
<br>Calculate the following:
<br>a. Max height above the ground.
<br>b. Total time in the air.
<br>c. Total distance traveled.
<br>d. Velocity vector at max height.
<br>e. Velocity when the ball hits the ground.

```python
from math import sin, radians, cos, sqrt

def sqr(x): # DIY Square Root Function
    a = (x**2)
    return(a)
# variable declaration
h = 1 #meter
angle = radians(30) #degree
v = 15 #m/s
g = 9.8 #m/s
#rad = radians(angle)
vix = vx = v*cos(angle)
viy = v*sin(angle)
vy = 0
V = sqrt(sqr(vx)+sqr(vy))
t = viy/g
H =(sqr(v)*sqr(sin(angle)))/(2*g)
T = (viy+sqrt(sqr(viy)+2+g+H))/g
max_h = h+H
d = vix*T
print("(a) Maximum height above the ground is {height} meter".format(height=max_h))
print("(b) Total time in the air is {T} second(s)".format(T=T))
print("(c) Total distance traveled is {d} meter".format(d=d))
print("(d) Velocity vector 'Vy' at Max height: {vy}\n    Velocity vector 'Vx' at Max height: {vx} m/s".format(vx=vx, vy=vy))
print("(e) Velocity when ball hits the ground is {V} m/s".format(V=V))
```
